```python
class CityWeather:
    def __init__(self, name, region, current, feels_like, condition):
        self.name = name
        self.region = region
        self.current_temp = current
        self.feels_like = feels_like
        self.condition = condition
    
    def __repr__(self):
        return f"<CityWeather|{self.city}>"
    
    def __str__(self):
        return f"It is currently {self.condition} and {self.current_temp} in {self.city}, {self.region}. It feels like {self.feels_like}."
        
```

```python
c1 = CityWeather('Chicago', 'Illinois', 44.6, 39.2, 'Clear')
print(c1)
c1
```

```python
class WeatherAPI:
    base_url = "http://api.weatherapi.com/v1"
    
    def __init__(self, api_key):
        self.api_key = api_key
      
        
    def __get(self, city, api_method, **params):
        request_url = f"{self.base_url}{api_method}?key={self.api_key}&q={city}"
        if params:
            for key, value in params.items():
                request_url += f"&{key}={value}"
        weather_response = requests.get(request_url)
        if weather_response.ok:
            return weather_response.json()
        else:
            print('There as an error!')

            
    def get_current_weather(self, city):
        weather_data = self.__get(city, '/current.json')
        city_name = weather_data['location']['name']
        print(city_name)
        
        region = weather_data['location']['region']
        print(region)
        
        current_temp = weather_data['current']['temp_f']
        print(current_temp)

        feels_like = weather_data['current']['feelslike_f']
        print(feels_like)

        current_condition = weather_data['current']['condition']['text']
        print(current_condition)

        
client = WeatherAPI(api_key)
client.get_current_weather('Chicago')
```

#### Pokemon API Wrapper

Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
!pip install requests
import requests
```

```python
base_url = "https://pokeapi.co/api/v2/"
api_method = "pokemon/"

pokemon = input('Enter the name of your Pokemon: ').lower()
api_url = f"{base_url}{api_method}{pokemon}"
print(api_url)
```

```python
poke_response = requests.get(api_url)  #<Response [200]>
poke_data = poke_response.json()
# poke_data  #prints the whole pokemon data
```

```python
name = poke_data['name']
order = poke_data['order']
poke_type = poke_data['types'][0]['type']['name']
weight = poke_data['weight']
height = poke_data['height']
```

```python
class PokemonInfo:
    
    def __init__(self):
        pass
    
    def get_stat(self):
        print(f"####################################")
        print(f"##            POKEDEX")
        print(f"####################################")
        print(f"##                                    ")
        print(f"## NAME: {name.upper()}")
        print(f"##                                    ")
        print(f"##      ORDER NO. #0{order}")
        print(f"##  POKEMON TYPE: {poke_type.title()} ")
        print(f"##        HEIGHT: {height}")
        print(f"##        WEIGHT: {weight}")
        print(f"##                                     ")
        print(f"#####################################")

pokemon1 = PokemonInfo()      
pokemon1.get_stat()
```
